Ex7: Add a new product to the inventory and ensure that its name and quantity are input by the user. 
Implement input validation to ensure that the quantity is a positive integer.

```python
def get_valid_quantity():
    while True:
        try:
            new_quantity=int(input("Enter the quantity of new item"))
            if new_quantity<=0:
                raise ValueError("Quantity must be positive integer")
            return new_quantity
        except ValueError as err:
            print(f"Invalid input. {err} try again")
        
new_item=input("Enter the name of new item")
new_quantity=get_valid_quantity()
inventory[new_item]=new_quantity
```

Ex9: Find and fix any variable misuse or logical errors in the code snippet provided.

```python
def reduce_item_stock(inventory, item_name, reduction_amount): 
    try: 
        new_amount = inventory[item_name] - reduction_amount 
        if new_amount >= 0: 
            inventory[item_name] = new_amount 
            print(f"The stock of {item_name} has been reduced by {reduction_amount}. New stock is {new_amount}.") 
        else: 
            print(f"Cannot reduce {item_name} by {reduction_amount}. Not enough stock.") 
    except KeyError: 
        print(f"{item_name} not found in the inventory.")
item_name=input("Enter item name to reduce stock level")
reduction_amount=int(input("Enter reduction amount for item"))
reduce_item_stock(inventory,item_name,reduction_amount)
```

```python
bookstore = [ 
    { 
        "isbn": "9781529434620", 
        "title": "The Art of Explanation", 
        "author": "Ros Atkins", 
        "stock": [ 
            {"location": "London", "quantity": 20}, 
            {"location": "Manchester", "quantity": 14}, 
            {"location": "Edinburgh", "quantity": 18}, 
            {"location": "Bristol", "quantity": 16}, 
            {"location": "Cardiff", "quantity": 15} 
        ],
        "price": 19.99 
    }, 
    {  "isbn": "9781529375992", 
        "title": "The Age of AI", 
        "author": "Henry a Kissinger", 
        "stock": [ 
            {"location": "London", "quantity": 10}, 
            {"location": "Birmingham", "quantity": 24}, 
            {"location": "Edinburgh", "quantity": 15}, 
            {"location": "Glasgow", "quantity": 14}, 
            {"location": "Belfast", "quantity": 19} 
        ], 
        "price": 10.99 
    }, 
    { 
        "isbn": "9781529434637", 
        "title": "Our Final Invention", 
        "author": "James Barrat", 
        "stock": [ 
            {"location": "Birmingham", "quantity": 5}, 
            {"location": "Liverpool", "quantity": 7}, 
            {"location": "Leeds", "quantity": 12}, 
            {"location": "Sheffield", "quantity": 11}, 
            {"location": "Newcastle", "quantity": 10} 
        ], 
        "price": 14.99 
    } 
] 

       
def restock_books(bookstore,stock_threshold,order_quantity):
    try:
        if not isinstance(bookstore,list) or not isinstance(stock_threshold,int) or not isinstance(order_quantity,int):
            raise ValueError("Invalid input types")
        if stock_threshold<0 or order_quantity<=0:
            raise ValueError("Stock value and order_quantity must be positive value.")
        for book in bookstore:
            for city_stock in book['stock']:
                if city_stock['quantity']<stock_threshold:
                    city_stock['quantity']+=order_quantity
        return bookstore
        
    except ValueError as err:
        print(f"Error,{err}")
                
bookstore=restock_books(bookstore,15,5)
for book in bookstore:
    for item,value in book.items():
        print(f"{item}:{value}")
```

```python
car = { 
    "model": "Mustang", 
    "manufacturer": "Ford", 
    "year": 1966, 
    "features": { 
        "colour": "blue", 
        "engine_type": "V8", 
        "is_convertible": True 
    }
}
```

```python
print(len(car))
```

```python
def count_properties(car): 
    property_count = len(car)  # Count the top-level properties 
 
    # Add the count of nested dictionary properties 
    for value in car.values():
        if isinstance(value, dict): 
            property_count += len(value) 
 
    return property_count
property_count=count_properties(car)
print(property_count)
```
