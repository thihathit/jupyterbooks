```python
import numpy as np
def read_csv_to_numpy(file_path):
    return np.genfromtxt(file_path, delimiter=',', skip_header=1)
def perform_analysis_numpy(number_of_stores):
    annual_sales = {}
    average_transaction_values = {}
    annual_gross_margins = {}
    for store_number in range(1, number_of_stores + 1):
        data = read_csv_to_numpy(f'chain_store_{store_number}_sales.csv')
        #print(data)
        annual_sales[store_number] = np.sum(data[:, 0])
        average_transaction_values[store_number] = np.mean(data[:, 1])
        annual_gross_margins[store_number] = np.sum(data[:, 2])
    highest_sales_store = max(annual_sales, key=annual_sales.get)
    lowest_sales_store = min(annual_sales, key=annual_sales.get)

    return annual_sales,average_transaction_values,annual_gross_margins,highest_sales_store,lowest_sales_store
result=perform_analysis_numpy(2)
print(result)
```
