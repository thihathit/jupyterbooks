calculate the sum and average across each row and each column.

```python
interesting_numbers=np.array([[1,2,3],[4,5,6]])
print(interesting_numbers)
sum_by_row = np.sum(interesting_numbers, axis=1)
print(sum_by_row)
sum_by_column = np.sum(interesting_numbers, axis=0) 
print(sum_by_column)
average_by_row = np.mean(interesting_numbers, axis=1)
print(average_by_row)
average_by_column = np.mean(interesting_numbers, axis=0) 
print(average_by_column)
```

Ex9: A class of 30 students have taken four programming tests, each graded out of 20 marks. You want 
to perform an analysis on the test results to understand the class’s performance. 
The test marks are stored in a 30x4 NumPy array, where each row represents a student and each 
column represents one of the four tests.

```python
test_marks = np.array([[10, 16, 11,  3], 
       [ 5, 11, 12,  9], 
       [12, 10,  1, 13], 
       [18,  9,  7,  1], 
       [15,  4, 14, 12], 
       [16, 17,  7,  6], 
       [ 4, 11,  8,  2], 
       [16,  1, 12, 14],
       [ 6, 13,  5,  4], 
       [ 3, 15,  9,  5], 
       [18, 15, 14,  8], 
       [ 2,  6,  7, 18], 
       [ 3,  2, 14, 19], 
       [18, 19, 12,  1], 
       [ 8,  2,  8, 14], 
       [14, 13,  9, 10], 
       [ 5,  8,  6,  8], 
       [14, 11,  8, 18], 
       [15,  7, 19, 12], 
       [ 4, 16, 19,  4], 
       [13, 15, 10,  7], 
       [17,  7, 14,  7], 
       [ 9, 18, 15,  5], 
       [ 8, 10, 17,  4], 
       [ 9, 19, 17,  1], 
       [19,  5,  8,  6], 
       [17, 11, 18,  1], 
       [19, 18,  6, 19], 
       [ 7, 11, 16,  3], 
       [ 2,  8, 15,  6]])
test_marks
```

Task 1 
Calculate the mean, median, and standard deviation for the marks of each test.  

```python
mean_marks=np.mean(test_marks,axis=0)
median_marks=np.median(test_marks,axis=0)
std_marks=np.std(test_marks,axis=0)
print("Average marks for each test",mean_marks)
print("Median mark for each test",median_marks)
print("Std mark for each test",std_marks)
```

Task 2 
Find the highest and lowest mark obtained in each test.

```python
highest_mark=np.max(test_marks,axis=0)
lowest_mark=np.min(test_marks,axis=0)
print("Highest mark in each test",highest_mark)
print("Lowest mark in each test",lowest_mark)
```

Task 3 
Identify the student with the highest total mark. 

```python
total_marks=np.sum(test_marks,axis=1)
print("Total marks for each student",total_marks)
```

```python
student_index=np.argmax(total_marks)
student_highest_mark=total_marks[student_index]
print("Highest total mark=",student_highest_mark)
```

Task 4 
The class has taken another test, graded out of 20. Add it to the array so that the array becomes 
30x5. 

```python
test_5 = np.array([ 
    [ 1], [ 7], [ 3], [ 4], [ 1], [18], [18], [ 9], [ 3], [15], 
    [16], [ 1], [ 5], [13], [ 4], [ 7], [ 2], [ 9], [ 5], [11], 
    [16], [ 9], [ 4], [11], [17], [13], [19], [ 1], [ 9], [ 1] 
]) 
test_5
```

```python
updated_mark=np.hstack((test_marks,test_5))
print("New Mark array=",updated_mark)
```

Task 5 
You realise that there was an error in the marking of the third test. All students are to receive an 
additional 2 marks for that test, but the total should not exceed the maximum of 20 marks.  
Update the marks of the third test and ensure they are capped at 20 marks. 
Hint: Use a NumPy function where().  

```python
updated_mark[:, 1] = np.where(updated_mark[:, 1] > 20, 20, updated_mark[:, 1]) + 2
updated_mark
```
