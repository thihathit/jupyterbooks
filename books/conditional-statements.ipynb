```python
def calculate_discount(age, purchase_amount):
    discount_rate = 0  # No discount

    if age >= 60 and purchase_amount >= 100:
        discount_rate = 0.15 # 15% discount
    elif 18 <= age <= 59 and purchase_amount >= 150:
        discount_rate = 0.10 # 10% discount
    elif age < 18 and purchase_amount >= 200:
        discount_rate = 0.05  # 5% discount

    # Calculate the final amount after applying the discount
    final_amount = purchase_amount * (1 - discount_rate)
    
    return final_amount, discount_rate
```

```python
 # Input from the user
age = int(input("Enter your age: "))
purchase_amount = float(input("Enter the purchase amount: "))

# Calculate the discount
final_amount, discount_rate = calculate_discount(age, purchase_amount)

# Output the result
if discount_rate > 0:
    print(f"You are eligible for a {discount_rate*100}% discount.")
    print(f"Your final amount after discount is: Â£{final_amount:.2f}")
else:
    print("Sorry, you are not eligible for any discounts.")
```
