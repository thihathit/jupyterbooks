Task 5 
add a new column 'Management Level' with three different levels 'Top' 'Middle' 'Operative, conditional on the column 'Grade' 
Grade 10 or above is 'Top', Grade 5 to 9 inclusive is 'Middle', and Grade 4 or under is 'Operative' 

```python
df['Management Level']=np.random.choice(['Top','Middle','Operative'],size=6)
df
```

```python
# Define a function to determine management level based on grade 
def determine_management_level(grade): 
    if grade >= 10: 
        return 'Top' 
    elif 5 <= grade <= 9: 
        return 'Middle' 
    else: 
        return 'Operative' 
 
# Apply the function to the 'Grade' column to create the new 'Management Level' column 
df['Management Level'] = df['Grade'].apply(determine_management_level)
df
```

Check out the updated DataFrame. 
Now it comes to the pay rise, anyone graded 6 or above receive a 2% salary rise and anyone 
graded 5 or below receives a 5% salary rise. Can you update it in the dataset?

```python
df['Salary'] = [50, 28, 68.4, 45.3, 80.2, 34]
df
```

```python
df['Salary'] = df.apply(lambda row: row['Salary'] * 1.02 if row['Grade'] >= 6 else row['Salary'] * 1.05, axis=1)
df
```
